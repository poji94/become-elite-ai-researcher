# 1. Simple Math Functions in Python

This notebook shows simple mathematical functions through examples.
A function is like a machine: you put in a number (x) and get out another number (y).


```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.random.mtrand import exponential
```

## 1. Linear Functions: y = x * k

The simplest function! Multiply x by a constant number k.


```python
# Example: y = 2 * x
# If x = 1, then y = 2
# If x = 2, then y = 4
# If x = 3, then y = 6

# k = 2 indicates the gradient or slope of function
# (through multiplication of x in defining value y)
# let's experiment a bit and change it to 3 or something
k = 5   # change here
x = 1
y = k * x
print(f"When x = {x}, y = {y}")

x = 5
y = k * x
print(f"When x = {x}, y = {y}")

# Let's plot it
x_values = np.linspace(0, 10, 100)
y_values = k * x_values

plt.figure(figsize=(8, 5))
plt.plot(x_values, y_values, 'b-', linewidth=2, label='y = 2x')
plt.xlabel('x')
plt.ylabel('y')
plt.title(f"Linear Function: y = {k} x")
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Different values of k
# positive values indicate positive slope - to the right side
# negative value indicates negative slope - to the left side
k_values = [1, 2, 3, 0.5, -1]
x = np.linspace(-5, 5, 100)

plt.figure(figsize=(10, 6))
for k in k_values:
    y = k * x
    plt.plot(x, y, label=f'y = {k}x', linewidth=2)

plt.xlabel('x')
plt.ylabel('y')
plt.title('Different Linear Functions: y = kx')
plt.grid(True)
plt.legend()
plt.axhline(y=0, color='k', linewidth=0.5)
plt.axvline(x=0, color='k', linewidth=0.5)
plt.show()
```

## 2. Quadratic Functions: y = x²

Square the input number. If x = 2, then y = 4. If x = 3, then y = 9.


```python
# Example: y = x²
examples = [-3, -2, -1, 0, 1, 2, 3]

# ** indicates operation for power (for example: ** 2 means power of 2 or square)
for x in examples:
    y = x ** 2
    print(f"When x = {x}, y = x² = {y}")

# Plot it
x_values = np.linspace(-5, 5, 100)
y_values = x_values ** 2

plt.figure(figsize=(8, 5))
plt.plot(x_values, y_values, 'r-', linewidth=2, label='y = x²')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Quadratic Function: y = x²')
plt.grid(True)
plt.legend()
plt.axhline(y=0, color='k', linewidth=0.5)
plt.axvline(x=0, color='k', linewidth=0.5)
plt.show()
```

## 3. Cubic Functions: y = x³

Cube the input number. If x = 2, then y = 8. If x = 3, then y = 27.


```python
# Example: y = x³
examples = [-2, -1, 0, 1, 2]
for x in examples:
    y = x ** 3
    print(f"When x = {x}, y = x³ = {y}")

# Plot it
x_values = np.linspace(-3, 3, 100)
y_values = x_values ** 3

plt.figure(figsize=(8, 5))
plt.plot(x_values, y_values, 'g-', linewidth=2, label='y = x³')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Cubic Function: y = x³')
plt.grid(True)
plt.legend()
plt.axhline(y=0, color='k', linewidth=0.5)
plt.axvline(x=0, color='k', linewidth=0.5)
plt.show()
```

## 4. Square Root Function: y = √x

The opposite of squaring. If x = 4, then y = 2. If x = 9, then y = 3.


```python
# Example: y = √x
examples = [0, 1, 4, 9, 16, 25]
for x in examples:
    y = np.sqrt(x)
    print(f"When x = {x}, y = √{x} = {y}")

# Plot it
x_values = np.linspace(0, 25, 100)
y_values = np.sqrt(x_values)

plt.figure(figsize=(8, 5))
plt.plot(x_values, y_values, 'm-', linewidth=2, label='y = √x')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Square Root Function: y = √x')
plt.grid(True)
plt.legend()
plt.show()
```

## 5. Exponential Function: y = 2^x

Double the result each time x increases by 1. If x = 1, then y = 2. If x = 2, then y = 4. If x = 3, then y = 8.


```python
# Example: y = 2^x
examples = [0, 1, 2, 3, 4, 5]
for x in examples:
    y = 2 ** x
    print(f"When x = {x}, y = 2^{x} = {y}")

# Plot it
x_values = np.linspace(-2, 5, 100)
y_values = 2 ** x_values

plt.figure(figsize=(8, 5))
plt.plot(x_values, y_values, 'c-', linewidth=2, label='y = 2^x')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Exponential Function: y = 2^x')
plt.grid(True)
plt.legend()
plt.axhline(y=0, color='k', linewidth=0.5)
plt.axvline(x=0, color='k', linewidth=0.5)
plt.show()
```

## 6. Sine Function: y = sin(x)

A wavy function that repeats. Used for waves, circles, and oscillations.


```python
# Example: y = sin(x)
# importing math library to use sin() function
import math

examples = [0, math.pi/2, math.pi, 3*math.pi/2, 2*math.pi]
for x in examples:
    y = math.sin(x)
    print(f"When x = {x:.2f}, y = sin(x) = {y:.2f}")

# Plot it
x_values = np.linspace(0, 4*np.pi, 100)
y_values = np.sin(x_values)

plt.figure(figsize=(10, 5))
plt.plot(x_values, y_values, 'orange', linewidth=2, label='y = sin(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Sine Function: y = sin(x)')
plt.grid(True)
plt.legend()
plt.axhline(y=0, color='k', linewidth=0.5)
plt.show()
```

## 7. Combined Functions: y = 2x + 3

You can combine operations! Multiply x by 2, then add 3.


```python
# Linear function with offset or known as (y-intercept)
# Example: y = 2x + 3
# set k as gradient or slope to be changeable for experimentation
# set b as offset or y-intercept to be changeable for experimentation
k = 5
b = -3

examples = [0, 1, 2, 3, 4]
for x in examples:
    y = 2 * k + b
    print(f"When x = {x}, y = 2({x}) + 3 = {y}")

# Plot it
x_values = np.linspace(-5, 5, 100)
y_values = k * x_values + b

plt.figure(figsize=(8, 5))
plt.plot(x_values, y_values, 'purple', linewidth=2, label=f'y = {k}x + {b}')
plt.xlabel('x')
plt.ylabel('y')
plt.title(f'Combined Function: y = {k}x + {b}')
plt.grid(True)
plt.legend()
plt.axhline(y=0, color='k', linewidth=0.5)
plt.axvline(x=0, color='k', linewidth=0.5)
plt.show()
```

## 8. Summary: Compare All Functions

Let's see how different functions look together!


```python
x = np.linspace(0, 5, 100)

plt.figure(figsize=(12, 8))

# Plot 1 (Linear function)
# changeable gradient for experimentation
gradient = 2
plt.subplot(2, 2, 1)
plt.plot(x, gradient * x, 'b-', linewidth=2)
plt.title(f'y = {gradient}x (Linear)')
plt.grid(True)

# Plot 2 (Quadratic)
plt.subplot(2, 2, 2)
plt.plot(x, x**2, 'r-', linewidth=2)
plt.title('y = x² (Quadratic)')
plt.grid(True)

# Plot 3 (Exponential)
# changeable slope for experimentation in exponential
plt.subplot(2, 2, 3)
plt.plot(x, 2 ** x, 'g-', linewidth=2)
plt.title('y = 2^x (Exponential)')
plt.grid(True)

# Plot 4 (Square root)
plt.subplot(2, 2, 4)
plt.plot(x, np.sqrt(x), 'm-', linewidth=2)
plt.title('y = √x (Square Root)')
plt.grid(True)

# draw all plot figures
plt.tight_layout()
plt.show()
```
